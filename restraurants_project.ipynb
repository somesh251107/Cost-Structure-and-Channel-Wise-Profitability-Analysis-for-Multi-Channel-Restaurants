```python
df["RestaurantID"]=df["RestaurantID"].astype(str)
```

```python
df["CuisineType"]=df["CuisineType"].astype("category")
```

```python
df["Segment"]=df["Segment"].astype("category")
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(df["AOV"])
plt.show()
```

```python
# PHASE 2 — Feature Engineering
```

```python
# 1. Total Revenue

df["TotalRevenue"]=(
    df["InStoreRevenue"]
    +df["UberEatsRevenue"]
    +df["DoorDashRevenue"]
    +df["SelfDeliveryRevenue"]
)
```

```python
df["TotalRevenue"]
```

```python
# 2. Total Net Profit

df["TotalNetProfit"]=(
    df["InStoreNetProfit"]
    +df["UberEatsNetProfit"]
    +df["DoorDashNetProfit"]
    +df["SelfDeliveryNetProfit"]
)
```

```python
df["TotalNetProfit"]
```

```python
# 3. Total Orders

df["TotalOrders"] = (
    df["InStoreOrders"]
    + df["UberEatsOrders"]
    + df["DoorDashOrders"]
    + df["SelfDeliveryOrders"]
)
```

```python
# 4. Profit Margin

df["ProfitMargin"]=(
    df["TotalNetProfit"]/df["TotalRevenue"]
)*100
```

```python
df["ProfitMargin"]
```

```python
df["TotalOrders"]
```

```python
# 5. Revenue Per Order

df["RevenuePerOrder"]=(
    df["TotalRevenue"]/df["TotalOrders"]
)
```

```python
# 6. Profit Per Order

df["ProfitPerOrder"]=(
    df["TotalNetProfit"]/df["TotalOrders"]
)
```

```python
# 7. Cost of Goods

df["COGS"]=(
    df["TotalRevenue"]*df["COGSRate"]
)
```

```python
# 8. Operating Expense

df["OperatingExpense"]=(
    df["TotalRevenue"]*df["OPEXRate"]
)
```

```python
# 9. Total Commission Cost

df["CommissionCost"]=(
    (df["UberEatsRevenue"]+df["DoorDashRevenue"])
    *df["CommissionRate"]
)
```

```python
# 10. Total Delivery Cost

df["TotalDeliveryCost"]=df["SD_DeliveryTotalCost"]
```

```python
# 11. Delivery Revenue

df["DeliveryRevenue"]=(
    df["UberEatsRevenue"]
    +df["DoorDashRevenue"]
    +df["SelfDeliveryRevenue"]
)
```

```python
# 12. Delivery Profit

df["DeliveryProfit"]=(
    df["UberEatsNetProfit"]
    +df["DoorDashNetProfit"]
    +df["SelfDeliveryNetProfit"]
)
```

```python
# 13. Delivery Margin

df["DeliveryMargin"]=(
    df["DeliveryProfit"]
    /df["DeliveryRevenue"]
)*100
```

```python
# 14. Commission Drag

df["CommissionDrag"]=(
    df["CommissionCost"]
    /df["TotalRevenue"]
)*100
```

```python
# 15. Self Delivery ROI

df["SelfDeliveryROI"]=(
    df["SelfDeliveryNetProfit"]
    /df["SD_DeliveryTotalCost"]
)
```
